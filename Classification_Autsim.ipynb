```python
#Lets convert rest of attributes/predictors in term of one hot encoding 
eth = OneHotEncoder().fit_transform(df[['ethnicity']]).toarray()
jund = OneHotEncoder().fit_transform(df[['jundice']]).toarray()
at = OneHotEncoder().fit_transform(df[['austim']]).toarray()
ctre= OneHotEncoder().fit_transform(df[['contry_of_res']]).toarray()

#Lets drop all original columns that are one hot encoded
df.drop(columns=['gender','ethnicity','jundice','austim','contry_of_res','Class/ASD'],inplace=True)
```

```python
#Lets convert all one hot encoded columns from Numpy Array to Data Frame
eDF=pd.DataFrame(eth).astype(int)
jDF=pd.DataFrame(jund).astype(int)
aDF=pd.DataFrame(at).astype(int)
cDF=pd.DataFrame(ctre).astype(int)
```

```python
#Lets combine all the datframe to retrive original X dataFrame or dataframe of all predictors
oneDF=pd.concat([df,genderDF, eDF,jDF,aDF,cDF], axis=1)
```

```python
X=oneDF.values
```

```python
"""
Dividing your data into test and train set is known as smapling. Here are types of sampling:
1- Random sampling
2- Smapling with replacement
3- Sampling without replacement
4- Stratified Sampling

Readings https://www.questionpro.com/blog/types-of-sampling-for-social-research/
"""
```

```python
#Train Test Split (It is called smapling)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
#Just want to look how my class labels look in Training set
y_train
```

```python
#Just want to see how my Training dataset look excluding class label
X_train
```

```python
#We are using tree based classifier for model building
#Reference reading:https://scikit-learn.org/stable/modules/tree.html
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

```python
y_predict=clf.predict(X_test)
```

```python
y_predict
```

```python
#Accuracy evaluation
#Rference reading https://www.kdnuggets.com/2020/05/model-evaluation-metrics-machine-learning.html
from sklearn import metrics
print("Train set Accuracy: ", metrics.accuracy_score(y_train, clf.predict(X_train)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test, y_predict))
```

```python
#Confusion matrix is another handy evaluation of model
#Readings https://towardsdatascience.com/the-5-classification-evaluation-metrics-you-must-know-aa97784ff226

from sklearn.metrics import confusion_matrix
labels=[0,1]
cm=confusion_matrix(y_test,y_predict,labels)
print(cm)
```

```python
"""
Confusion matrix:
            Actual 0.  Actual 1
Predicted 0 [94          9]
Predicted 1 [ 5         33]

Accuracy= [Predicted 0 and Actual 0] # + [Predicted 1 and Actual 1] # / Total popullation
        = 94+33/139
Error= [Predicted 1 but Actual 0] # + [Predicted 0 and Actual 1] # / Total popullation
     = 5+9/139
Precision= [Predicted 0 and Actual 0]/[Predicted 0 and Actual 0]+[Predicted 0 and Actual 1]
         = 94/94+9
Recall= [Predicted 0 and Actual 0]/[Predicted 0 and Actual 0]+[Predicted 1 and Actual 0]
      = 94/94+5
Type I error= [Predicted 0 and Actual 1]= 9
Type II error= [Predicted 1 and Actual 0]=5
"""
```

```python
#K nearest neighbor (K-NN) based classifier
#Readings https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html
#REadings https://towardsdatascience.com/machine-learning-basics-with-the-k-nearest-neighbors-algorithm-6a6e71d01761

from sklearn.neighbors import KNeighborsClassifier
k = 4
#Train Model and Predict  
neigh = KNeighborsClassifier(n_neighbors = k).fit(X_train,y_train)
neigh
```

```python
#Predicting
yhat = neigh.predict(X_test)
yhat[0:5]
```

```python
#Accuracy evaluation
from sklearn import metrics
print("Train set Accuracy: ", metrics.accuracy_score(y_train, neigh.predict(X_train)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test, yhat))
```

```python
#Confusion matrix is another handy evaluation of model
from sklearn.metrics import confusion_matrix
labels=[0,1]
cm=confusion_matrix(y_test,yhat,labels)
print(cm)
```
